```python
df2 = pd.read_csv('grayBreastCellLines_public_grayBreastCellLineExon_genomicMatrix.csv')
```

```python
df2.head()
```

```python
df2.describe()
```

```python
df2.info()
```

```python
df2.set_index('probes', inplace = True)
df2
```

```python
tumor_df = df2.replace(0.000000,np.nan)
tumor_df
```

```python
tumor_df.isnull()
```

```python
# Find the length of the shorter dataset
min_len = min(len(normal_df), len(tumor_df))

# Truncate the longer dataset to the length of the shorter dataset
normal_df_short = normal_df.head(min_len)
tumor_df_short = tumor_df.head(min_len)

print(normal_df_short)
print(tumor_df_short)
```

```python
# Concatenate along axis=1 (side by side)
merged_df = pd.concat([normal_df_short, tumor_df_short], axis=1)
```

```python
print("Merged dataset:")
print(merged_df)
```

```python
merged_df.describe()
```

```python
merged_data = merged_df.replace(0.000000,np.nan)
merged_data
```

```python
merged_data.isnull()
```

```python
merged_data.isnull().sum().sum()
```

```python
merged_data = merged_data.fillna(merged_data.mean(numeric_only = True))
merged_data
```

```python
# Calculate the correlation matrix
corr_matrix = merged_data.corr()
```

```python
# Print the correlation matrix
print(corr_matrix)
```

```python
# Find the correlation between two specific columns
col1 = 'K102416'
col2 = 'ZR7530'

correlation = merged_df[col1].corr(merged_df[col2])

print(f"Correlation between {col1} and {col2}: {correlation:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Visualize the correlation matrix using a heatmap
plt.figure(figsize=(30,28))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
from lifelines import KaplanMeierFitter
from lifelines import CoxPHFitter
```

```python
#Create a Kaplan-Meier plot
kmf=KaplanMeierFitter()
kmf.fit(merged_data['K102416'], event_observed=merged_data['SUM225CWN'])
```

```python
#Plot the Kaplan-Meier curve
plt.plot(kmf.survival_function_)
plt.xlabel('Time')
plt.ylabel('Survival Probability')
plt.title('Kaplan-Meier Curve')
plt.show()
```

```python
#Perform the Cox regression analysis
cph = CoxPHFitter()
cph.fit(merged_data[['K102416', 'SUM225CWN']], duration_col= 'K102416', event_col= 'SUM225CWN')
```

```python
#Prrint the results
cph.print_summary()
```

```python
#Plot the Cox regression curve
plt.plot(cph.baseline_survival_)
plt.xlabel('Time')
plt.ylabel('Survival Probability')
plt.title('Cox Regression Curve')
plt.show()
```
